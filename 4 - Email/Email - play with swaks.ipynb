# swaks

swaks is a Linux command that emulates an SMTP client. It sends an email to a specified email address, with a lot of default options that need not be specified explicitly: sender address, value of the HELO argument, mail body, mail headers, port number of the mail server, which kind of SMTP authentication to use and so on.

In most cases, one has to specify only very few options for overriding the default. Thus, a very compact command suffices to execute a complex sequence of actions.

By default, swaks connects to the SMTP server specified by the destination address; uses plain SMTP (i.e., no TLS); does not use any authentication. Essentially, it behaves as a mail server that sends emails to the mail server of another mail domain.

What is useful, from our point of view, is the fact that the command output displays the full SMTP interaction with the specified mail server.

The Linux virtual machine associated with this notebook runs in an environment without any limitation on the outbound SMTP traffic, i.e., it is possible to connect to any mail server in the world (via SMTP). In this execution environment, it is useful to always specify the *--helo* option and the *--from* option.



The following command sends a default email (i.e., default body and default headers) to the specified destination.

You may want to check what has been delivered at your DEST_ADDRESS.





```python
!swaks --from "carlo.ancelotti@realmadrid.es" --to DEST_ADDRESS --helo colab.com
```
